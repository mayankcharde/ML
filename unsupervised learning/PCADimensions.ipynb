```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
     
```

```python
X, y = make_blobs(n_samples=500, n_features=5, centers=3, cluster_std=1.5, random_state=42)

     
```

```python
#  THIS IS BASICALLY A 5 DIMENSION 
#  WHICH WE WANT TO CINVERT INTO 2 DIMENSIONS
X
```

```python


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
#  HERE WE ARE CONVERTING INTO DATAFRAMES 
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
df_pca['label'] = y
```

```python
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='label', palette='Set2')
```
